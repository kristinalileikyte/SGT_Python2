# 2. **SQL Grouping:**

### Task: Calculate the total sales amount for each product in the "sales" table.

```python
sales_product_amount = pd.read_sql_query("""

SELECT p.product_name, s.product_id, COUNT(sale_id) AS sales_amount
FROM sales AS s
INNER JOIN products AS p
    ON s.product_id = p.product_id
GROUP BY product_id;

""",con=engine)
sales_product_amount
```

### Task: Find the number of customers in each city from the "customers" table.

```python
customers_from_country = pd.read_sql_query("""

SELECT city, COUNT(*) AS total_customers
FROM customers 
GROUP BY city;

""",con=engine)
customers_from_country
```

### Task: Calculate the average salary for each department in the "employees" table.

```python
department_average_salary = pd.read_sql_query("""

SELECT department, AVG(salary) AS avg_salary
FROM employees 
GROUP BY department
ORDER BY salary;

""",con=engine)
department_average_salary
```

### Task: List the highest and lowest prices for each category in the "products" table.

```python
products_highest_price = pd.read_sql_query("""

SELECT category, MIN(price) AS min_price, MAX(price) AS max_price  
FROM products 
GROUP BY category;

""",con=engine)
products_highest_price
```

### Task: Find the number of orders placed by each customer in the "orders" table.

```python
customers_orders_number = pd.read_sql_query("""

SELECT c.customer_name, c.customer_id, COUNT(order_id) AS total_orders
FROM orders as o
INNER JOIN customers AS c
    ON o.customer_id = c.customer_id
GROUP BY customer_id;

""",con=engine)
customers_orders_number
```

### Task: Calculate the running total sales for each month in the "sales" table.

```python
sales_total_month = pd.read_sql_query("""

SELECT 
    case 
        WHEN MONTH(sale_date) = 4 THEN "April"
        WHEN MONTH(sale_date) = 5 THEN "May"
        WHEN MONTH(sale_date) = 6 THEN "June"
        WHEN MONTH(sale_date) = 7 THEN "July"
        WHEN MONTH(sale_date) = 8 THEN "August"
    end as month, 
    SUM(revenue) as total_sales 
FROM sales 
GROUP BY month
ORDER BY sale_date;

""",con=engine)
sales_total_month
```

### Task: Determine the average sales amount over a 3-month sliding window for each product in the "sales" table.

```python
sales_3month_average = pd.read_sql_query("""

SELECT p.product_name, s.product_id, 
    case 
        WHEN MONTH(sale_date) = 4 THEN "April"
        WHEN MONTH(sale_date) = 5 THEN "May"
        WHEN MONTH(sale_date) = 6 THEN "June"
        WHEN MONTH(sale_date) = 7 THEN "July"
        WHEN MONTH(sale_date) = 8 THEN "August"
    end as month,
    SUM(quantity) as total_quantity, ROUND(
    AVG(SUM(quantity)) OVER (PARTITION BY s.product_id ORDER BY sale_date ROWS BETWEEN 2 PRECEDING AND CURRENT ROW), 1) 
    AS 3_month_avg 
FROM sales AS s
INNER JOIN products AS p
    ON s.product_id = p.product_id
GROUP BY s.product_id,month
ORDER BY s.product_id,sale_date
;

""",con=engine)
sales_3month_average
```

### Task: Rank customers based on their total order amounts in the "orders" table.

```python
orders_customers_rank = pd.read_sql_query("""

SELECT c.customer_name, o.customer_id, COUNT(o.order_id) AS total_orders,
    DENSE_RANK() OVER (ORDER BY total_orders DESC) AS rank
FROM orders AS o
INNER JOIN customers AS c
    ON o.customer_id = c.customer_id
GROUP BY customer_id
ORDER BY rank;

""",con=engine)
orders_customers_rank
```

# **Window Functions in SQL:**

###  Task: Calculate the cumulative sum of revenue for each product in the "sales" table.

```python
sales_revenue_cumul_sum = pd.read_sql_query("""

SELECT p.product_name, s.product_id, s.sale_id, s.revenue,
    SUM(revenue) OVER (PARTITION BY product_id ORDER BY product_id ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) 
AS cumulative_sum
FROM sales AS s
INNER JOIN products AS p
    ON s.product_id = p.product_id;

""",con=engine)
sales_revenue_cumul_sum
```

### Task: Assign percentile ranks to employees based on their salaries in the "employees" table.

```python
employees_salary_ranks_percent = pd.read_sql_query("""

SELECT employee_name, salary,
    (PERCENT_RANK() OVER (ORDER BY salary)) AS percentile_rank
FROM employees
ORDER BY salary DESC;

""",con=engine)
employees_salary_ranks_percent
```

### Task: Retrieve the previous salary for each employee in the "employees" table, considering 0 as the default value for the first row.

```python
employees_previous_salary = pd.read_sql_query("""

WITH cte as (
  SELECT *,
    COALESCE(
      LAG(salary,1) OVER (ORDER BY employee_id), 0
    ) AS previous_salary 
  FROM employees
)
SELECT employee_name, salary, previous_salary
from cte;


""",con=engine)
employees_previous_salary
```

### Task: Calculate the sum of revenue for each product within a 2-day window, including the current row, in the "sales" table.

```python
sales_product_revenue_2days = pd.read_sql_query("""

SELECT p.product_name, s1.product_id, s1.sale_date, s1.revenue, 
    (
        SELECT SUM(revenue) 
        FROM sales AS s2
        WHERE (s2.sale_date BETWEEN s1.sale_date - interval 1 day AND s1.sale_date) 
            AND s1.product_id = s2.product_id
    ) AS revenue_2day
FROM sales AS s1
INNER JOIN products AS p
    ON s1.product_id = p.product_id
ORDER BY product_id, s1.sale_date;

""",con=engine)
sales_product_revenue_2days
```
